```python
def count_tuples(listA):
  unique_items = list(set(listA))
  item_count = [listA.count(item) for item in unique_items]
  return unique_items,item_count
```

```python
def count_lott_p_sure_amt_amb(data):
  df = pd.read_csv(data)
  df['crdm_trial_resp.corr'] = 1.0 - df['crdm_trial_resp.corr']
  df = df.loc[df['crdm_trial_type'] == 'task']
  
  sure_amt_count_df = df['crdm_sure_amt'].value_counts().sort_index()
  sure_amt_count_df = sure_amt_count_df.reset_index()
  sure_amt_count_df.rename(columns={"index":"sure_amt", "crdm_sure_amt":"frequency"}, inplace=True)
  print(sure_amt_count_df)
  fn = join(filepath,"CRDM","sure_amt.csv")
  print("Saving to: {}".format(fn))
  sure_amt_count_df.to_csv(fn)
  
  amb_lev_count_df = df['crdm_amb_lev'].value_counts().sort_index()
  amb_lev_count_df = amb_lev_count_df.reset_index()
  amb_lev_count_df.rename(columns={"index":"amb_lev", "crdm_amb_lev":"frequency"}, inplace=True)
  print(amb_lev_count_df)
  fn = join(filepath,"CRDM","amb_lev.csv")
  print("Saving to: {}".format(fn))
  amb_lev_count_df.to_csv(fn)

  lott_p_count_df = df['crdm_lott_p'].value_counts().sort_index()
  lott_p_count_df = lott_p_count_df.reset_index()
  lott_p_count_df.rename(columns={"index":"lott_p", "crdm_lott_p":"frequency"}, inplace=True)
  print(lott_p_count_df)
  fn = join(filepath,"CRDM","lott_ps.csv")
  print("Saving to: {}".format(fn))
  lott_p_count_df.to_csv(fn)
```

```python
def get_subject(data):
  subj = os.path.basename(data).replace(".csv","")
  return subj
```

```python
#get set of all good data files for analysis
filepath = "/content/drive/MyDrive/CloudResearchTrialDataApprovedFiles/"
good_files = sorted(glob(join(filepath, "idm-v2_2023-*")))
if (not good_files):
  print("No good files available. Check file path.")
  sys.exit()

fn_helper = join(filepath, 'idm-v2_2023-01-25_12h21.15.616.csv')
count_lott_p_sure_amt_amb(fn_helper)
```

Use the results from above to define upper and lower bounds and ylabels and titles

```python
lott_p_lower_bound = [0,20,30,40,60]
lott_p_upper_bound = [20,30,40,60,80]
amb_lower_bound = [-1, 20, 40, 60]
amb_upper_bound = [20, 40, 60, 80]
ylabels = ['Amb 0.0','Amb 24.0', 'Amb 50.0', 'Amb 74.0']
titles = ['Lott 13.0p', 'Lott 25.0p','Lott 38.0p','Lott 50.0p','Lott 75.0p']

cols = ["lott_p", "amb_level", "sure_amt","lott_amt_choice", "frequency"]
lott_amt_choice_df = pd.DataFrame(columns=cols)

for i, data in enumerate(good_files):
  print(data)
  df = pd.read_csv(data)
  df['crdm_trial_resp.corr'] = 1.0 - df['crdm_trial_resp.corr']
  df['crdm_lott_amt'] = df['crdm_lott_top'] + df['crdm_lott_bot']
  df = df.loc[df['crdm_trial_type'] == 'task']

  plt.figure(i,figsize=(20,10))  

  #loop through each lottery-probability for each ambuiguity level
  for iimmed, (il,iu) in enumerate(zip(amb_lower_bound,amb_upper_bound)):
    for iplot,(lb,ub) in enumerate(zip(lott_p_lower_bound,lott_p_upper_bound)):
      index = 1+iplot+(5*iimmed)
      plt.subplot(4,5,index)
      x = []
      y = []
          
      idf = df.loc[(df['crdm_lott_p'] < ub) & (df['crdm_lott_p'] > lb) & (df['crdm_amb_lev'] < iu) & (df['crdm_amb_lev'] > il)]
      x = x + idf['crdm_lott_amt'].tolist()
      y = y + idf['crdm_trial_resp.corr'].tolist()
      
      lott_amt_choice, count = count_tuples(list(zip(x,y)))
      this_df = pd.DataFrame(columns=cols)
      this_df["frequency"] = count
      this_df["lott_amt_choice"] = lott_amt_choice
      this_df["lott_p"] = titles[iplot]
      this_df["sure_amt"] = ylabels[iimmed]
      lott_amt_choice_df = lott_amt_choice_df.append(this_df,ignore_index=True)


      if (not x):
        pass
      else:
        x,y = zip(*sorted(zip(x,y)))
      plt.plot(x,y,'*-')
      plt.ylim([-0.1,1.1])
      plt.xlim([0,df['crdm_lott_amt'].max()])
      if (iplot == 0):
        plt.ylabel(ylabels[iimmed],fontsize=15)
      if (iimmed == 0):
        plt.title(titles[iplot],fontsize=15)
      if (iimmed == 3):
        plt.xlabel("Lottery Amount",fontsize=15)
  
  subj = get_subject(data)
  subj_dir = os.path.join(filepath,"CRDM",subj)
  if (not os.path.exists(subj_dir)):
    os.makedirs(subj_dir)

  fn = join(filepath,"CRDM", subj, "lott_amt_choice.csv")
  print("Saving to: {}".format(fn))
  lott_amt_choice_df.to_csv(fn)
  print(lott_amt_choice_df)
  plt.suptitle(subj, fontsize=25)
  fig_fn = join(filepath,"CRDM",subj, "crdm_plot_lott_amt_choice.png")
  print("Saving to: {}".format(fig_fn))
  plt.savefig(fig_fn)
  plt.show()
```
